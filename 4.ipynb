# 4.5 Scatterplot

```r
examData = read.delim('data/Exam Anxiety.dat', header=TRUE)
```

```r
scatter = ggplot(examData, aes(Anxiety, Exam))
```

```r
scatter + geom_point()
```

```r
scatter + geom_point() + labs(x='Exam Anxiety', y='Exam Performance %')
```

```r
scatter + geom_point() + geom_smooth() + labs(x='Exam Anxiety', y='Exam Performance %')
```

```r
scatter + geom_point() + geom_smooth(method='lm', color='Red', alpha=0.1, fill='Blue') + labs(x='Exam Anxiety', y='Exam Performance %')
```

```r
scatter = ggplot(examData, aes(Anxiety, Exam, color=Gender))
```

```r
scatter + geom_point() + geom_smooth(method='lm', aes(fill=Gender), alpha=0.1) + labs(x='Exam Anxiety', y='Exam Performance %')
```
